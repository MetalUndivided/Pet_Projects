# Bird Classifier

Built using the Kaggle sourced <a href=https://www.kaggle.com/datasets/gpiosenka/100-bird-species>dataset</a> of 525 bird species.

## Data Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
df = pd.read_csv('/media/metal_undivided/DATA/Datasets/Kaggle_525_species/birds.csv')
```

Let's see what's inside the csv.

```python
display(df.head())
```

Data labels are clear, now to reading the images.

```python
datagen = ImageDataGenerator(rescale=1/255)
```

```python
seed = 111 ## for reproducibility
```

```python
train_flow = datagen.flow_from_dataframe(
    df[df['data set'] == 'train'],
    directory='/media/metal_undivided/DATA/Datasets/Kaggle_525_species/',
    x_col='filepaths',
    y_col='class id',
    target_size=(224, 224),
    class_mode='raw',
    seed=seed
)
```

```python
valid_flow = datagen.flow_from_dataframe(
    df[df['data set'] == 'valid'],
    directory='/media/metal_undivided/DATA/Datasets/Kaggle_525_species/',
    x_col='filepaths',
    y_col='class id',
    target_size=(224, 224),
    class_mode='raw',
    seed=seed
)
```

```python
test_flow = datagen.flow_from_dataframe(
    df[df['data set'] == 'test'],
    directory='/media/metal_undivided/DATA/Datasets/Kaggle_525_species/',
    x_col='filepaths',
    y_col='class id',
    target_size=(224, 224),
    class_mode='raw',
    seed=seed
)
```

Now that the data is loaded, let's inspect it.

## Data Inspection

```python
label_map = dict(zip(df['class id'], df['labels']))
```

```python
batch = next(train_flow)
```

```python
fig, axs = plt.subplots(4, 4, figsize=(20, 20))

pixels = batch[0]
label = batch[1]

for i in range(16):
    axobject = axs[i // 4][i % 4]
    axobject.imshow(pixels[i])
    axobject.set_title(label_map[label[i]])
    axobject.get_xaxis().set_visible(False)
    axobject.get_yaxis().set_visible(False)
```

Looks like the data is properly imported and is ready for training.
